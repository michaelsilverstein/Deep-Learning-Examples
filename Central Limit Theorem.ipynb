# Central Limit Theorem

The central limit theorem states that the sum of a set of random variables becomes increasingly Gaussian as the number of variables increases. With 100 random variables, we can illustrate this concept with increasing dimensionality of each random variable.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10,6)
```

```python
fig,axs = plt.subplots(3,3)
axs = axs.ravel()
for ax,i in zip(axs,range(1,10)):
    N = i**3
    ax.hist(np.sum(np.random.rand(100,N),axis=1))
    ax.set_title('N = %d'%N)
plt.tight_layout()
plt.show()
```
